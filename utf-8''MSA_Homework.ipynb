The following code illustrates how to perform measurement tests and capability test in the R language using the "SixSigma" library.

```r
# Install the "SixSigma" library.
install.packages('SixSigma')
```

```r
# Load the "SixSigma" package.
library(SixSigma)
```

### MEASUREMENT TEST

```r
# Load the measurement data.
msa_data <- read.csv('http://bit.ly/gage_data')
```

```r
# Preview the first few records of measurement data.
head(msa_data)
```

```r
# Adjust the default size of the plot to make it readable.
options(repr.plot.width=8, repr.plot.height=8)

# Perform a Gage R&R measurement test.
ss.rr(var = Measurement,
      part = Part,
      appr = Operator,
      data = msa_data)
```

### CAPABILITY STUDY

```r
# Load the sample data.
capability_data <- read.csv('http://bit.ly/cable_wires')
```

```r
# Preview the first few records of measurement data.
head(capability_data)
```

```r
# Adjust the default size of the plot to make it readable.
options(repr.plot.width=8, repr.plot.height=8)

# Perform a capability assessment.
ss.study.ca(capability_data$Diameter,
            LSL = 0.5, 
            USL = 0.6, 
            Target = .55, 
            alpha = 0.5)
```
